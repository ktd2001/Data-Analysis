```python

import numpy as np
import matplotlib.pyplot as plt
import math

# Values for X and Y
x = np.array([0,1,2,3,4,5])
y = np.array([10,9,6,4,6,10])

plt.plot(x,y)
```

```python
# Slope equation is m = (y2 - y1) / (x2 - x1)
Change_in_y = (y[2] - y[1])
print(Change_in_y)
```

```python
Change_in_x = (x[2] - x[1])
print(Change_in_x)
```

```python
m =  (Change_in_y) / (Change_in_x)
print(m)
# The slope is -3 between X= 1 and X= 2 and descending just as the curve is descending.
```
